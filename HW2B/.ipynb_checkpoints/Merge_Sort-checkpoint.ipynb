```python
#Merge Sort
#Recursively Merge Sort to get O(log(n)
#left = left list of list A(returned in base case of merge_sort) 
#right= right list of list A(returned in base case of merge_sort)
def algorithm_merge(left, right, left_count, right_count): 
    #next line:  merges the # of flips of left array and right array
    count = left_count + right_count #add the left_count and right_count
    i = 0#index of left list
    j = 0#index of right list
    C = []#create new list
    #Iterate through lists until both of the index are more then the size of both list    
    #while loop sorts left list and right list--concatenated into new list C
    while i <len(left) and j <len(right):
        if left[i] <= right[j]:#If the element in list A is less then the element in list B
            C.append(left[i])#append element if true
            i = i + 1
        else: #appends right[j] to C, counts number of flips if right[j] > left[i]
            C.append(right[j])#If not append element in list B, aka the correct order
            j = j + 1
            new= len(left) - i# number of flips to sort right[j] into new list C
            count = count+ new
    #if any of the list are different sizes append the rest into our new list
    C = C + left[i:]
    C = C + right[j:]
    return C, count #return sorted array C and count = number of flips used to sort array C

def algorithm_merge_sort(A, count):
    # Note: once we reach our max recurse depth we will beging to go back up our tree
    if len(A) <=1: 
        return A, count #return count, count = 0, if len(A) = 1 
    mid = len(A) // 2 #floor divde length
    #split array A in the next two lines
    #variable "left" = left half of the list
    #variable "right" = right half of the list
    ##Merge_sort returns left_count = number of flips to sort left half of list A
    left, left_count = algorithm_merge_sort(A[:mid], count)
    ##merge sort returns right_count = number of flips to sort right half of list A
    right, right_count  = algorithm_merge_sort(A[mid:], count)
    #Merge left and right side, input left_count and right_count
    return algorithm_merge(left, right, left_count,right_count)

A = [1, 20, 6, 4, 5]
count = 0

print(algorithm_merge_sort(A,count))
```

```python
#To assign carrots to rabbits and minimize the time it takes to eat the last carrot we use 
#QuickSort uses random pivoting to achieve the average case consistently of nlog(n)

import random
rabbits = [7,3,2,13,2]
carrots = [1,3,5,14,21]

def Randomized_Quicksort(A,p,r):
    if p < r:
        q,A = RPartition(A, p, r)
        return Randomized_Quicksort(A,p,q-1)
        return Randomized_Quicksort(A,q+1,r)
    else:
        return A
    
  
def RPartition(A,p,r):
    i = random.randrange(p,r)#choose uniformly random integer on [p..r]
    temp = A[i]
    A[i] = A[r]
    A[r] = temp
    x = A[r]
    i = p - 1
    j = p
    for j in range(p, r):
        if A[j] <= x:
            i = i+1
            temp = A[i]
            A[i] = A[j]
            A[j] = temp  
    temp = A[i+1]
    A[i+1] = A[r]
    A[r] = temp
    return i + 1, A
 
A = Randomized_Quicksort(rabbits, 0, len(rabbits)-1) # sort rabbits
B = Randomized_Quicksort(carrots, 0, len(carrots)-1) # sort carrots
print("A sorted: ",A)
print("B sorted: ",B)
counter = 0
ans = []
for i in range(0, len(A)):#find difference 
    diff = A[i] - B[i]
    ans.append(abs(diff))
    counter = counter + 1
print(max(ans))#print max of in the array to return the longest time
    
```
